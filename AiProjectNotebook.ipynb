## Min max normalization:
For data transformation we applied the min max normalization to rescale the data, min max normalization guarantees all features have the exact same scale.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data=pd.DataFrame(scaler.fit_transform(data),
            columns=data.columns, index=data.index) 
print(data.head(4))
```

```python
# Visual Clarification of our dataset after applying the normaliztion: 
sns.jointplot(data = data , dropna = True)
plt.show()
```
